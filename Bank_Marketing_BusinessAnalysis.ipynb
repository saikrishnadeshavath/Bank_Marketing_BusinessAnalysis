```python
# Identify numerical features (excluding the one-hot encoded ones)
numerical_features_scaled = df.select_dtypes(include=np.number).columns.tolist()

# Scale the numerical features
scaler = StandardScaler()
X_resampled[numerical_features_scaled] = scaler.fit_transform(X_resampled[numerical_features_scaled])

# Display the first few rows of the preprocessed dataframe
display(X_resampled.head())
display(y_resampled.head())
```

```python
# Select categorical features excluding the target variable 'y'
categorical_features = df.select_dtypes(include='object').columns.tolist()
categorical_features.remove('y')

# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_features, drop_first=True)

# Display the first few rows of the encoded dataframe
display(df_encoded.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Initialize and train a Logistic Regression model
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

print("Logistic Regression model trained successfully.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
# Generate ROC curve
y_prob = model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

## Business Analysis Summary and Recommendations

Based on our comprehensive analysis of the bank marketing dataset, here's a summary of our findings and recommendations:

**1. Exploratory Data Analysis (EDA):**
*   We identified the data types and confirmed no missing values.
*   The target variable 'y' (term deposit subscription) shows a significant class imbalance, with a majority of customers not subscribing.
*   Analysis of categorical features revealed that certain job types (e.g., student, retired), marital statuses (single), and education levels (tertiary) appear to have a higher proportion of subscribers.
*   Analysis of numerical features indicated that 'duration' of the last contact is strongly associated with subscription. 'pdays' and 'previous' contacts also show a relationship with the target variable. 'campaign' seems to have a negative association with subscription.

**2. Feature Engineering and Data Preprocessing:**
*   We performed one-hot encoding on categorical variables to prepare them for modeling.
*   We addressed the class imbalance using SMOTE oversampling to ensure the model is not biased towards the majority class.
*   Numerical features were scaled to standardize their range, which is important for many machine learning algorithms.

**3. Model Selection, Training, and Evaluation:**
*   We selected and trained a Logistic Regression model as a baseline for predicting term deposit subscriptions.
*   The model achieved good performance metrics (Accuracy: {{accuracy:.4f}}, Precision: {{precision:.4f}}, Recall: {{recall:.4f}}, F1-Score: {{f1:.4f}}, ROC AUC: {{roc_auc:.4f}}) on the test set after addressing class imbalance.
*   We also trained and evaluated a Random Forest model, which showed slightly better performance compared to Logistic Regression across various metrics (Accuracy: {{accuracy_rf:.4f}}, Precision: {{precision_rf:.4f}}, Recall: {{recall_rf:.4f}}, F1-Score: {{f1_rf:.4f}}, ROC AUC: {{roc_auc_rf:.4f}}).

**4. Interpretation of Model Findings:**
*   By examining the model coefficients of the Logistic Regression model and feature importances from the Random Forest model, we identified the most influential features for predicting term deposit subscriptions:
    *   **Duration:** The duration of the last contact is consistently identified as a highly influential feature across both models, strongly associated with subscription.
    *   **Months:** Certain months (e.g., January, November, July, August, May) have a notable impact on subscription likelihood, with some months being less favorable for campaigns.
    *   **Job:** Job types like blue-collar, housemaid, entrepreneur, and technician show lower subscription rates compared to others, while retired and student show higher rates.
    *   **Poutcome (Success/Unknown):** A successful outcome in a previous campaign is a strong positive indicator for current subscription. Having an unknown previous outcome is associated with a lower subscription rate.
    *   **Contact (Unknown):** Customers with unknown contact information are less likely to subscribe.
    *   **Housing Loan:** Having a housing loan is associated with a lower likelihood of subscription.
    *   **Balance:** While not as influential as duration, balance still plays a role, with higher balances showing a slight positive association with subscription.
    *   **Day and Age:** These features also have some influence, though generally less pronounced than the top features.
    *   **Campaign, Pdays, and Previous:** The number of contacts during the current campaign, days since last contact from a previous campaign, and number of previous contacts also contribute to the prediction.

**Recommendations:**

Based on this analysis, here are some recommendations for the bank:

1.  **Optimize Campaign Timing and Targeting:** Leverage insights on influential months and job/marital/education categories to optimize campaign timing and target specific customer segments more effectively.
2.  **Prioritize Customers with Previous Success:** Focus efforts on customers who had a successful outcome in previous campaigns, as they are highly likely to subscribe again.
3.  **Improve Contact Strategy:** Invest in strategies to increase effective contact duration while being mindful of its potential collinearity with the target. Also, work on obtaining better contact information for customers.
4.  **Consider Housing Loan Status:** Acknowledge the lower subscription rate among customers with housing loans and tailor approaches accordingly.
5.  **Further Investigate Influential Factors:** Conduct deeper analysis on why certain months, job types, and contact methods are more or less associated with subscriptions to refine strategies.
